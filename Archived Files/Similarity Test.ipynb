Clustering Using K-Means

```python
df = pd.read_csv('testSongData.csv')
songlist = df.values.tolist()
```

```python
for i in range(0, len(songlist)):
    del(songlist[i][0])
```

```python
for i in range(0, len(songlist)):
    del(songlist[i][0])
```

```python
for i in range(0, len(songlist)):
    del(songlist[i][0])
```

```python
model = sklearn.cluster.KMeans(n_clusters=24)
```

```python
labels = model.fit_predict(songlist)
```
